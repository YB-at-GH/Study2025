```python
# Complete OSM Representation - City Name

from shapely.geometry import Polygon
import osmnx as ox
import matplotlib.pyplot as plt
import networkx as nx
import osmnx as ox
import numpy as np

city = 'palo alto, california, united states'

GG = ox.graph.graph_from_place(city, network_type='drive', simplify=True, retain_all=False, truncate_by_edge=False, which_result=None, custom_filter=None)

# Basic stats
Gs = ox.stats.basic_stats(GG, area=None, clean_int_tol=None)
display(Gs)

# Get all edge lengths
edge_lengths = [d['length'] for u, v, k, d in GG.edges(data=True, keys=True)]

# Calculate median
median_length = np.median(edge_lengths)
print(f"Median street length: {median_length:.2f} meters")

# Plot the road network
fig, ax = plt.subplots(figsize=(12, 12))
ax.set_facecolor('black')
ox.plot_graph(GG, ax=ax, node_size=0, edge_color='gray', edge_linewidth=0.5, show=False)
plt.show()
```

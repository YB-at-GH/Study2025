```python
# Core-Amsterdam Polygon
import osmnx as ox
from shapely.geometry import Polygon
import matplotlib.pyplot as plt
import networkx as nx
import osmnx as ox
import numpy as np

# Define your polygon
Amsterdam_Core_Polygon = Polygon([
    (4.780528861265865, 52.3884653052481),
    (4.807475465138122, 52.33161653707547),
    (4.908729370597514, 52.3266263134687),
    (4.956906632066095, 52.341595295552324),
    (4.967521960864257, 52.35406557753191),
    (4.97650416215501, 52.37625397663567),
    (4.902196860567876, 52.38373070116879),
    (4.876475102326175, 52.40690050239024),
    (4.839321451532608, 52.39145531988559),
    (4.89920446601881, 52.425751178414146),
    (4.780528861265865, 52.3884653052481),
])

# Get the street network within the polygon
GPL = ox.graph_from_polygon(Amsterdam_Core_Polygon, network_type='bike')


# Basic stats
Gs = ox.stats.basic_stats(GPL, area=None, clean_int_tol=None)
display(Gs)

# Get all edge lengths
edge_lengths = [d['length'] for u, v, k, d in GPL.edges(data=True, keys=True)]

# Calculate median
median_length = np.median(edge_lengths)
print(f"Median street length: {median_length:.2f} meters")

# Plot the road network
fig, ax = plt.subplots(figsize=(12, 12))
ax.set_facecolor('black')
ox.plot_graph(GPL, ax=ax, node_size=0, edge_color='gray', edge_linewidth=0.5, show=False)
plt.show()
```
